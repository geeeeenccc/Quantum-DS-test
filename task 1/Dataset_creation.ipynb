## Labeling

With get_markers function we'll be able to automatically label the dataset, marking the spans within the tweets that correspond to mountain names as entities.

```python
# Sample DataFrame with a 'text' column
data = df

# this function will return a list of markers for the given text and list of names
# of the mountains
# Basically, it will return a list of tuples of the form (start_index, end_index)
def get_markers(text, mountain_names):
    # Initialize an empty list to store the markers
    markers = []

    # Iterate over all mountain names except "Alps"
    for mountain_name in mountain_names:
        if mountain_name != "Alps":
            # Check if the mountain name is present in the text
            if mountain_name in text:
                # If it is, add the start and end indices of the mountain name to the markers list
                start_index = text.find(mountain_name)
                end_index = start_index + len(mountain_name)
                markers.append((start_index, end_index))

    # Check if "Alps" is present in the text
    if "Alps" in text:
        # If it is, add the start and end indices of "Alps" to the markers list
        start_index = text.find("Alps")
        end_index = start_index + len("Alps")
        markers.append((start_index, end_index))

    # Return the markers list
    return markers

# List of mountain names
mountain_names = [
    "Denali", "Dolomites", "Carpathian Mountains", "Mount Olympus",
    "Sierra Nevada mountains", "Tien Shan", "Tatra Mountains", "Blue Ridge Mountains",
    "Mount Cook", "Alps", "Mount Everest", "Himalayas", "Drakensberg Mountains",
    "Kunlun Mountains", "Taurus Mountains", "Rhodope Mountains",
    "Mount Rainier", "Cascade Range", "Mount Kilimanjaro",
    "Andes", "Rocky Mountains", "Atlas Mountains",
    "Mount Elbrus", "Appalachian Mountains", "Mount Fuji",
    "Dolomite Mountains", "Ural Mountains", "Patagonian Andes", "Columbia Icefield",
    "Brooks Range", "Mount Eiger", "Great Dividing Range", "Mount St. Helens",
    "Tian Shan Mountains", "Apennine Mountains", "Rwenzori Mountains", "Mount Logan",
    "Mount Kenya", "Scandinavian Mountains", "Sierra Madre Occidental", "K2",
]
# Apply the function to the 'text' column and create a new 'marker' column
data['marker'] = data['text'].apply(lambda text: get_markers(text, mountain_names))

# Display the updated DataFrame
data
```

```python
# This function will divide a single marker on multiple based on number of words in the marker
# For example:
# The Apennine Mountains ... [(4, 22)]
# The marker will be devided into [(4, 12), (13, 22)]

def divide_markers(text, markers):
    divided_markers = []

    for start, end in markers:
        # Extract the mountain name from the text based on the marker
        mountain_name = text[start:end]

        # Split the mountain name into individual words
        words = mountain_name.split(" ")

        # Generate divided markers for each word
        for word in words:
            word_start = text.find(word, start)
            word_end = word_start + len(word)
            divided_markers.append((word_start, word_end))

    return divided_markers


# Apply the divide_markers function to the DataFrame
df['markers'] = df.apply(lambda row: divide_markers(row['text'], row['marker']), axis=1)

# Display the updated DataFrame
df
```
